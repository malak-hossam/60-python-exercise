**32-write a python program to find the Strongest Neighbour (Input: 1 2 2 3 4 5
Output: 2 2 3 4 5)

```python
x = [1,2,3,4,5]
xx = []
for i in range(len(x)):
    a = x[i-1] if i > 0 else None
    b = x[i+1] if i < len(x)-1 else None
    aa = max(a,b) if a is not None and b is not None else a or b
    xx.append(aa)
print(xx)
```
